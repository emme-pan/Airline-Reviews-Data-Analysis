```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
## Initialize dataframe
df = pd.DataFrame()
list = ["Aircraft", "Type of Traveller", "Seat Type", "Date Flown", "Seat Comfort", "Cabin Staff Service", "Food & Beverages", 
        "Inflight Entertainment", "Ground Service", "Wifi & Connectivity", "Value For Money", "Recommended", "Date of Post",
        "Overall Rating (/10)"]

for i in list:
    df[i] = ""

#df["Overall Rating (/10)"] = ""

#df["Aircraft"] = ""
df
```

```python
# Lists:
type_of_traveller_list = ['Business', 'Family Leisure', 'Couple Leisure', 'Solo Leisure']
type_of_seat_list = ['First Class', 'Business Class', 'Premium Economy', 'Economy Class']
month_list = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October','November',
             'December']
list_text_headers = ['Aircraft', 'Type Of Traveller', 'Seat Type', 'Route', 'Date Flown']

# String for star rating name:
# "<span class=\"star fill\">1</span>"
string1 = "<span class=\"star fill\">"
string2 = "</span>"

# -----------------------------------------------------------------------------------------------------------------------
# Website Stuff:
# -----------------------------------------------------------------------------------------------------------------------
base_url = "https://www.airlinequality.com/airline-reviews/air-canada"
#pages = 2
page_size = 100
page_counter = 0

page = requests.get(base_url) 
    
soup = BeautifulSoup(page.content, "html.parser")
results = soup.find(id="main")

# Total number of pages:
top = soup.find(id="main")
#total_reviews = results.find("div", class_ = "pagination-total")
total_reviews = top.find("span", itemprop = "reviewCount").text.strip()
total_pages = np.ceil(int(total_reviews) / page_size)
pages = int(total_pages)

## Page:
results = soup.find(id="main")
total_reviews = results.find("div", class_ = "pagination-total")

for j in range(1, pages + 1):
    #print("Page: " + str(page_counter))
    page_counter = page_counter + 1
    
    url = f"{base_url}/page/{j}/?sortby=post_date%3ADesc&pagesize={page_size}"
    page = requests.get(url) 
    
    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find(id="main")
    
    review_elements = results.find_all("div", class_="tc_mobile")
    overall_rating_elements = results.find_all("div", class_="rating-10")
    datepost_elements = results.find_all("time", itemprop="datePublished")
    
    # -----------------------------------------------------------------------------------------------------------------------
    # Start of Loop for DataFrame:
    # ---------------------------------------------------------------------------------------------------------------------
    for i in range(len(review_elements)): 
        # -----------------------------------------------------------------------------------------------------------------
        # Text Reviews:
        # ----------------------------------------------------------------------------------------------------------------- 
        # Aircraft:---------------------------------------------------------------------
        if review_elements[i].find_all("td", class_="review-rating-header aircraft"):
            aircraft_element = review_elements[i].find("td", class_="review-value")
            data_aircraft = {"Aircraft": [aircraft_element.text.strip()]}
        else:
            data_aircraft = {"Aircraft": [None]}
            
        # Type of Traveller:----------------------------------------------------------------
        if review_elements[i].find_all("td", class_="review-rating-header type_of_traveller"):
            type_traveller_elements = review_elements[i].find_all("td", class_="review-value")
            for type_traveller_element in type_traveller_elements:
                for traveller_type in type_of_traveller_list:
                    if traveller_type in type_traveller_element:
                        data_type_traveller = {"Type of Traveller": [traveller_type]}
        else:
            data_type_traveller = {"Type of Traveller": [None]}
            
        # Type of Seat:----------------------------------------------------------------
        if review_elements[i].find_all("td", class_="review-rating-header cabin_flown"):
            type_seat_elements = review_elements[i].find_all("td", class_="review-value")
            for type_seat_element in type_seat_elements:
                for seat in type_of_seat_list:
                    if seat in type_seat_element:
                        data_seat_type = {"Seat Type": [seat]}
        else:
            data_seat_type = {"Seat Type": [None]}
        
        # Date Flown:----------------------------------------------------------------
        if review_elements[i].find_all("td", class_="review-rating-header date_flown"):
            date_flown_elements = review_elements[i].find_all("td", class_="review-value")
            for date_flown_element in date_flown_elements:
                for month in month_list:
                    if month in date_flown_element.text:
                        data_date_flown = {"Date Flown": [date_flown_element.text.strip()]}
        else:
            data_date_flown = {"Date Flown": [None]}
    
        # ----------------------------------------------------------------------------------------------------------------
        # Star Rating Reviews:
        # ----------------------------------------------------------------------------------------------------------------     
        # Loop for all the star ratings in the review and create a list per review:
        list_of_all_star_ratings_per_review = []
        star_elements = review_elements[i].find_all("td", class_="review-rating-stars stars")
        for star_element in star_elements:
            list_of_all_star_ratings_per_review.append(star_element)
        
        # Create list of all ratings completed by the reviewer:
        list_headers = []
        header_elements = review_elements[i].find_all("td", class_="review-rating-header")
        for header in header_elements:
            list_headers.append(header.text.strip())
        
        # Remove from the list, the text headers and keep only the star rating header names
        for header in list_text_headers:
            if header in list_headers:
                list_headers.remove(header)
        
        # Acquire all of the star ratings:--------------------------------------------------------------------------
        star_elements = review_elements[i].find_all("td", class_="review-rating-stars stars")
        
        # Loop for all the star ratings in the review and create a list per review:
        list_of_all_star_ratings_per_review = []
        for star_element in star_elements:
            list_of_all_star_ratings_per_review.append(star_element)
        
        # ----------------------------------------------------------------------------------------
        # Seat Comfort Rating:----------------------------------------------------------------------
        # ----------------------------------------------------------------------------------------
        star_counter = 0
    
        if review_elements[i].find_all("td", class_="review-rating-header seat_comfort"):
            star_elements = review_elements[i].find("td", class_="review-rating-stars stars")
            counter = 1
            for star_element in star_elements:
                if str(star_element) == string1 + str(counter) + string2:
                    star_counter = star_counter + 1
                counter = counter + 1
            data_seat_comfort = {"Seat Comfort": [star_counter]}
        else:
            counter = 0
            data_seat_comfort = {"Seat Comfort": None}
            
        # ----------------------------------------------------------------------------------------
        # Cabin Staff Service Rating:
        # ----------------------------------------------------------------------------------------
        star_counter = 0
        if 'Cabin Staff Service' in list_headers:
            index = list_headers.index('Cabin Staff Service')
        else:
            index = None
    
        if 'Cabin Staff Service' in list_headers:
            list_star_rating_cabin_staff = list_of_all_star_ratings_per_review[index]
            # make a loop to count how many stars:
            counter = 1 # Counter for 1 - 5
            for element in list_star_rating_cabin_staff:
            #print(element)
                if str(element) == string1 + str(counter) + string2:
                    star_counter = star_counter + 1
                    counter = counter + 1
            data_cabin_staff = {"Cabin Staff Service": [star_counter]}   
        else:
            star_counter = None
            list_star_rating_cabin_staff = None
            data_cabin_staff = {"Cabin Staff Service": [star_counter]}
            
        # ----------------------------------------------------------------------------------------
        # Food and Beverages Rating:
        # ----------------------------------------------------------------------------------------
        star_counter = 0
        if 'Food & Beverages' in list_headers:
            index = list_headers.index('Food & Beverages')
        else:
            index = None
     
        if 'Food & Beverages' in list_headers:
            list_star_rating_food_bev = list_of_all_star_ratings_per_review[index]
            # make a loop to count how many stars:
            counter = 1 # Counter for 1 - 5
            for element in list_star_rating_food_bev:
            #print(element)
                if str(element) == string1 + str(counter) + string2:
                    star_counter = star_counter + 1
                counter = counter + 1
            data_food_bev = {"Food & Beverages": [star_counter]}   
        else:
            star_counter = None
            list_star_rating_food_bev = None
            data_food_bev = {"Food & Beverages": [star_counter]}
        
        # ----------------------------------------------------------------------------------------
        # Inflight Entertainment Rating:
        # ----------------------------------------------------------------------------------------
        star_counter = 0
        if 'Inflight Entertainment' in list_headers:
            index = list_headers.index('Inflight Entertainment')
        else:
            index = None
        # 
        if 'Inflight Entertainment' in list_headers:
            list_star_rating_inflight_ent = list_of_all_star_ratings_per_review[index]
            # make a loop to count how many stars:
            counter = 1 # Counter for 1 - 5
            for element in list_star_rating_inflight_ent:
            #print(element)
                if str(element) == string1 + str(counter) + string2:
                    star_counter = star_counter + 1
                counter = counter + 1
            data_inflight_ent = {"Inflight Entertainment": [star_counter]}   
        else:
            star_counter = None
            list_star_rating_inflight_ent = None
            data_inflight_ent = {"Inflight Entertainment": [star_counter]}
        
        # ----------------------------------------------------------------------------------------
        # Ground Service Rating:
        # ----------------------------------------------------------------------------------------
        star_counter = 0
        if 'Ground Service' in list_headers:
            index = list_headers.index('Ground Service')
        else:
            index = None
     
        if 'Ground Service' in list_headers:
            list_star_rating_ground_service = list_of_all_star_ratings_per_review[index]
            # make a loop to count how many stars:
            counter = 1 # Counter for 1 - 5
            for element in list_star_rating_ground_service:
            #print(element)
                if str(element) == string1 + str(counter) + string2:
                    star_counter = star_counter + 1
                counter = counter + 1
            data_ground_service = {"Ground Service": [star_counter]}   
        else:
            star_counter = None
            list_star_rating_ground_service = None
            data_ground_service = {"Ground Service": [star_counter]}
            
        # ----------------------------------------------------------------------------------------
        # Wifi & Connectivity Rating:
        # ----------------------------------------------------------------------------------------
        star_counter = 0
        if 'Wifi & Connectivity' in list_headers:
            index = list_headers.index('Wifi & Connectivity')
        else:
            index = None
        # 
        if 'Wifi & Connectivity' in list_headers:
            list_star_rating_wifi = list_of_all_star_ratings_per_review[index]
            # make a loop to count how many stars:
            counter = 1 # Counter for 1 - 5
            for element in list_star_rating_wifi:
            #print(element)
                if str(element) == string1 + str(counter) + string2:
                    star_counter = star_counter + 1
                counter = counter + 1
            data_wifi = {"Wifi & Connectivity": [star_counter]}   
        else:
            star_counter = None
            list_star_rating_wifi = None
            data_wifi = {"Wifi & Connectivity": [star_counter]}
            
        # ----------------------------------------------------------------------------------------
        # Value For Money Rating:
        # ----------------------------------------------------------------------------------------
        star_counter = 0
        if 'Value For Money' in list_headers:
            index = list_headers.index('Value For Money')
        else:
            index = None
        # 
        if 'Value For Money' in list_headers:
            list_star_rating_money = list_of_all_star_ratings_per_review[index]
            # make a loop to count how many stars:
            counter = 1 # Counter for 1 - 5
            for element in list_star_rating_money:
            #print(element)
                if str(element) == string1 + str(counter) + string2:
                    star_counter = star_counter + 1
                counter = counter + 1
            data_value_money = {"Value For Money": [star_counter]}   
        else:
            star_counter = None
            list_star_rating_money = None
            data_value_money = {"Value For Money": [star_counter]}
        
        # ----------------------------------------------------------------------------------------
        # Recommended Rating:
        # ---------------------------------------------------------------------------------------- 
        if review_elements[i].find_all("td", class_="review-rating-header recommended"):
            recommended_element_yes = review_elements[i].find("td", class_="review-value rating-yes")
            recommended_element_no = review_elements[i].find_all("td", class_="review-value rating-no")
            if recommended_element_yes:
                data_recommended = {"Recommended": ["Yes"]}
            elif recommended_element_no:
                data_recommended = {"Recommended": ["No"]}
            else:
                data_recommended = {"Recommended": [None]}
                
        # ----------------------------------------------------------------------------------------
        # Date of Post:
        # ----------------------------------------------------------------------------------------
        data_post_date = {"Date of Post": [datepost_elements[i].text.strip()]}
    
        # ----------------------------------------------------------------------------------------
        # Overall Rating:
        # ----------------------------------------------------------------------------------------
        overall_rating = overall_rating_elements[i+1].find("span", itemprop="ratingValue")
        if overall_rating:
            data_overall_rating = {"Overall Rating (/10)": [overall_rating.text.strip()]}
        else: 
            data_overall_rating = {"Overall Rating (/10)": [None]}
        
        # -------------------------------------------------------------------------------------------------------------
        # Create final DataFrame:-------------------------------------------------------------
        # -------------------------------------------------------------------------------------------------------------
        data = (data_aircraft | data_type_traveller | data_seat_type | data_date_flown | data_seat_comfort | data_cabin_staff | 
                data_food_bev | data_inflight_ent | data_ground_service | data_wifi | data_value_money | data_recommended |
                data_post_date | data_overall_rating)    
        temp_df = pd.DataFrame.from_dict(data)
        df = pd.concat([df,temp_df], ignore_index = True)

    
df
```
